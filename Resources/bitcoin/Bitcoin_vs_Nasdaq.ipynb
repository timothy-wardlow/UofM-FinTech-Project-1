```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
bitcoin = pd.read_csv(Path("bitcoin_cleaned.csv"), header=0)
bitcoin.head()
```

```python
bitcoin.hvplot("Date")
```

```python
nasdaq_11=pd.read_csv(Path("nasdaq_11_cleaned.csv"), header=0)
nasdaq_11.head()
```

```python
nasdaq_11.hvplot("Date")
```

```python
bitcoin.hvplot("Date")*nasdaq_11.hvplot("Date")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
bitcoin.plot(x='Date', y = 'Bitcoin', ax=ax,ylabel='Bitcoin in US $')
nasdaq_11.plot(y='Nasdaq', ax=ax,secondary_y=True).set_ylabel('Nasdaq in US $')
```
