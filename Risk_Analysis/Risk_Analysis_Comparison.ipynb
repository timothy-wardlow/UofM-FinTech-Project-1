```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Read in NASDAQ Composite 1995-2005 data from csv file
IXIC_path = Path('^IXIC.csv')
IXIC_df = pd.read_csv(IXIC_path, parse_dates=True, infer_datetime_format=True, index_col='Date')
IXIC_df
```

```python
# Checking for nulls
IXIC_df.isnull().sum()
```

```python
# Dropping additional columns
IXIC_df = IXIC_df.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'])
IXIC_df
```

```python
IXIC_df.plot(figsize=(20,10), ylabel='IXIC 1995-2005 Closing Price')
```

```python
# Reading in Nasdaq Composite 2014-2021 data
IXIC_2_path = Path('^IXIC_2.csv')
IXIC_2_df = pd.read_csv(IXIC_2_path, parse_dates=True, infer_datetime_format=True, index_col='Date')
IXIC_2_df
```

```python
# Checking for nulls
IXIC_2_df.isnull().sum()
```

```python
# Dropping additional columns
IXIC_2_df = IXIC_2_df.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'])
IXIC_2_df
```

```python
IXIC_2_df.plot(figsize=(20,10), ylabel='IXIC 2011-2021 Closing Price')
```

```python
# Read in BTC 2011-2021 data from csv file
BTC_path = Path('BTC_USD.csv')
BTC_df = pd.read_csv(BTC_path, parse_dates=True, infer_datetime_format=True, index_col='Date')
BTC_df
```

```python
# Checking for nulls
BTC_df.isnull().sum()
```

```python
# Dropping additional columns
BTC_df = BTC_df.drop(columns=['Open', 'High', 'Low', 'Adj Close', 'Volume'])
BTC_df
```

```python
BTC_df.plot(figsize=(20,10), ylabel='BTC 2014-2021 Price')
```

```python
# Calculate daily returns for NASDAQ Composite from 1995-2005 and removing null values
IXIC_daily_returns = IXIC_df.pct_change()
IXIC_daily_returns.dropna(inplace=True)
IXIC_daily_returns.isnull().sum()
```

```python
IXIC_daily_returns
```

```python
IXIC_daily_returns.columns = ['^IXIC Returns']
IXIC_daily_returns.head()
```

```python
IXIC_daily_returns.plot(figsize=(20,10), ylabel='^IXIC 1995-2005 Returns')
```

```python
# Calculate daily returns for NASDAQ Composite from 2014-2021 and removing null values
IXIC_2_daily_returns = IXIC_2_df.pct_change()
IXIC_2_daily_returns.dropna(inplace=True)
IXIC_2_daily_returns.isnull().sum()
```

```python
IXIC_2_daily_returns
```

```python
IXIC_2_daily_returns.columns = ['^IXIC 2 Returns']
IXIC_2_daily_returns.head()
```

```python
IXIC_2_daily_returns.plot(figsize=(20,10), ylabel= 'IXIC 2011-2021 Returns')
```

```python
# Calculating BTC daily returns and removing null values
BTC_daily_returns = BTC_df.pct_change()
BTC_daily_returns.dropna(inplace=True)
BTC_daily_returns.isnull().sum()
```

```python
BTC_daily_returns
```

```python
BTC_daily_returns.columns = ['BTC Returns']
BTC_daily_returns.head()
```

```python
BTC_daily_returns.plot(figsize=(20,10), ylabel='BTC 2014-2021 Returns')
```

```python
# Joining dataframes
joined_returns = pd.concat([BTC_daily_returns, IXIC_daily_returns, IXIC_2_daily_returns])
joined_returns
```

```python
# Plotting returns
joined_returns.plot(figsize=(20,10), ylabel='Returns')
```

```python
# IXIC standard deviation
IXIC_std = IXIC_daily_returns.std()
IXIC_std
```

```python
IXIC_2_std = IXIC_2_daily_returns.std()
IXIC_2_std
```

```python
# BTC standard deviation
BTC_std = BTC_daily_returns.std()
BTC_std
```

```python
#Annualized standard deviation
IXIC_annual_std = IXIC_std * np.sqrt(252)
IXIC_2_annual_std = IXIC_2_std * np.sqrt(252)
BTC_annual_std = BTC_std * np.sqrt(252)
print('The annualized standard deviation for the Nasdaq Composite from 1995-2005 is: ' , IXIC_annual_std)
print('The annualized standard deviation for the Nasdaq Composite from 2014-2021 is: ' , IXIC_2_annual_std)
print('The annualized standard deviation for BTC from 2014-2021 is: ', BTC_annual_std)
```

```python
joined_annual_std = pd.concat([IXIC_annual_std, IXIC_2_annual_std, BTC_annual_std])
joined_annual_std.plot(kind='bar',figsize=(20,10), ylabel='Annualized Standard Deviation')
```

```python
sharpe_ratio = ((joined_returns.mean())*252) / (joined_returns.std() * np.sqrt(252))
sharpe_ratio
```

```python
sharpe_ratio.plot(kind='bar', figsize=(20,10), ylabel='Sharpe Ratio', title='Sharpe Ratios of BTC 2014-2021, IXIC 1995-2005, IXIC 2011-2021')
```
